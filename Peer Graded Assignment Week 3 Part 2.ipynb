### STEP 2 - Read Dataset from URL

```python
# Read url
post_codes_canada = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')[0]
print(post_codes_canada)
```

### STEP 3 - Convert html data into a Pandas DataFrame

```python
# Convert HTML data into dataframe, ignoring not assigned boroughs
df = post_codes_canada[post_codes_canada.Borough != 'Not assigned']
df.reset_index(drop = True)
print(df.shape)
df.head()
```

As instructed, Im taking the geographical coordinates of each postal code from given url since I wasn't able to get them using the Geocoder package

```python
url = 'http://cocl.us/Geospatial_data'
```

```python
import pandas as pd
df_geo_coord = pd.read_csv(url)
df_geo_coord.head()
```

In order to get the requested dataframe we need to join the Borough and the Neighbourhood from dataframe "df" in Part I with its respective geographical coordinates recently obtained and stored in dataframe "df_geo_coord"

```python
combined_df=df.join(df_geo_coord.set_index('Postal Code'), on='Postal Code')
combined_df.groupby(['Postal Code']).first().reset_index()
```
